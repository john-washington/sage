## SIR+ model

## This is a working model from the class and using the MIT open source code to adapt to our model

(Reworked from http://epirecip.es/epicookbook/chapters/sir/julia)

- $S(t)$ = number of *s*usceptible individuals at time $t$
- $I(t)$ = number of *i*nfectious individuals at time $t$
- $R(t)$ = number of *r*ecovered / removed individuals at time $t$
- $D(t)$ = number of *diseased/death* individuals at time $t$

Satisfy $S + I + R - D = N$, total population

Write in terms of the *proportions* $S(t) / N$.

At *each* time $t > 0$ we have the following **system of ODEs**:

\begin{align*}
\frac{dS(t)}{dt} &= -\beta S(t) I(t) \\
\frac{dI(t)}{dt} &= +\beta S(t) I(t) - \gamma I(t) - \mu I(t)\\
\frac{dR(t)}{dt} &= +\gamma I(t) \\
\frac{dD(t)}{dt} &= \mu I(t) \\
I_{n+1} &= I_n + \mu I_n \\
\lambda &= 1 + \mu \\
I_{(n+1)} &= \lambda I_n \\
I' &= -(S' + R' + D') \\
\gamma + \mu &= 1
\end{align*}

$dS / dt$ is the *rate of change* of $S(t)$

Simpler: Rewrite without the explicit functions of $t$, and use notation $\dot{S} := \frac{dS}{dt}$

\begin{align*}
\dot{S} &= -\beta S I \\
\dot{I} &= +\beta S I - \gamma I - \mu I\\ 
\dot{R} &= +\gamma I \\
\dot{D} &= \mu I \\
I_{n+1} &= I_n + \mu I_n \\
\lambda &= 1 + \mu \\
I_{(n+1)} &= \lambda I_n \\
I' &= -(S' + R' + D') \\
\gamma + \mu &= 1
\end{align*}

We define a function to represent the right-hand side of the system of ODEs.

We need a function that accepts the vector `u` of the current values of all the variables $u = (S(t), I(t), R(t), D(t))$, as well as the time $t$ and the parameter values $p$.

It also accepts another variable `du`. This is another (pre-allocated) vector, where it will write the output. Note that you should *not* create a new vector yourself; you must write into the pre-existing vector `du`.

You must use this ordering of the arguments to the function in order to use that function with `DifferentialEquations.jl`.

```julia
using Pkg
Pkg.add("DifferentialEquations")
using DifferentialEquations
```

$\mathbf{u}(t) = (S(t), I(t), R(t), D(t))$

```julia
function sir_ode!(du, u, p, t)  # ! since modifies du -- in-place updating (can have better performance)
    
    # unpack variables and parameters:
    S, I, R, D = u
    β, γ, μ = p   # \beta<TAB>
    
    # define differential equations:
    dS = -β*S*I
    dI = +β*S*I - γ*I -μ * I
    dR = +γ*I
    dD = +μ * I
    du .= (dS, dI, dR, dD)   # copy the values into the vector du; note the `.`
end
```

```julia
β = 0.00001
γ = 1/14
μ = 1/100

parameters = [β, γ, μ]

S₀ = 45400   # write subscript 0 as S\_0<TAB>
I₀ = 2040
R₀ = 2500
D₀ = 0

initial_values = [S₀, I₀, R₀, D₀]

time_span = [0.0, 200.0]  # initial and final time

# set up problem:
problem = ODEProblem(sir_ode!, initial_values, time_span, parameters)
```

```julia
typeof(ODEProblem)
```

```julia
# solve problem:
solution = solve(problem, saveat = 0.1) 
```

```julia
typeof(solution)
```

```julia
using Plots
```

```julia
plot(solution, label=["S" "I" "R" "D"])  # row vector of labels
```

## Calculating with uncertainties

People have created many surprising packages in Julia. For example, `Measurements.jl` allows us to calculate with **numbers with uncertainties**:

```julia
using Pkg
Pkg.add("Measurements")
using Measurements
```

```julia
x = 0.1 ± 0.01   # \pm<TAB>
```

```julia
2x
```

```julia
x^2
```

```julia
typeof(x)
```

We can think of this as saying "we are uncertain about the value of $x$: it is somewhere in the range (interval) $[0.1 - 0.01, 0.1 + 0.01]$"

Any measurement of a real physical, biological, economic etc. system has an uncertainty like this.

## SIR model with uncertainty

One of the amazing things about Julia is the way different packages can be used together: we call this **composability**. 

For example, can we solve an SIR model with uncertainty?

In order to do so, the packages must be written in such a way that they **do not assume anything (or much) about the *types* of their inputs**. 

```julia
β = 0.1 ± 0.01
γ = 0.05 ± 0.01
μ = 0.01 ± 0.01

parameters = [β, γ, μ]

S₀ = 0.99   # write subscript 0 as S\_0<TAB>
I₀ = 0.01 ± 0.001
R₀ = 0.0
D₀ = 0.0 ± 0.001

initial_values = [S₀, I₀, R₀, D₀]

t_span = [0.0, 200.0]  # initial and final time

# set up problem:
problem = ODEProblem(sir_ode!, initial_values, t_span, parameters)

# solve problem:
solution = solve(problem, saveat = 1.0)
```

```julia
plot(solution, alpha=0.5, label=["S" "I" "R" "D"])
```

## Our own measurements type

```julia
struct MyMeasurement
    mid::Float64
    width::Float64
end
```

```julia
 MyMeasurement(3, 1)
```

```julia
function Base.show(io::IO, x::MyMeasurement)
    print(io, "$(x.mid) ± $(x.width)")
end
```

```julia
x = MyMeasurement(3, 1)
```

```julia
Measurements.:±(x, y) = MyMeasurement(x, y)
```

```julia
import Base: +

+(x::MyMeasurement, y::MyMeasurement) = MyMeasurement(x.mid + y.mid,
                                                      x.width + y.width)
```

```julia
x = MyMeasurement(2, 0.1)
y = MyMeasurement(3, 0.01)
```

```julia
x + y
```

```julia
@edit 3 ± 4
```

```julia
@edit (3 ± 4) + 3
```

## Anonymous functions

What is an **anonymous function**?

Looks like

```julia
x -> 3x
```

"The function that maps a value x to the value 3x"

```julia
(x -> 3x)(10)
```

```julia
g = x -> 3x
```

```julia
g(10)
```

In Julia: Common to pass functions as arguments to other functions.
A function that takes another function as argument is a **higher-order function**

Useful to anonymous functions

E.g. `map`:

```julia
f(x) = 2x
```

```julia
v = [1, 2, 3]
```

```julia
f.(v)
```

```julia
map(f, v)
```

```julia
map(x -> 2x+1, v)
```
