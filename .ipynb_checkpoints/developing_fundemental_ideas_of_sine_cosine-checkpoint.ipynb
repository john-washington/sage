**The Power Series expansion formula for sine and cosine**

can be developed after observing and reasoning about the restoring force, calculated with the double integral, as follows:

$\sin(x) = \sum_{n=0}^\infty \frac{(-1)^n x^{2n+1}}{(2n +1)!}$

and 

$\cos(x) = \sum_{n=0}^\infty \frac{(-1)^n x^{2n}}{(2n)!}$

```python
def my_sine(x):
    n=var('n')
    return sum((-1)^n*(x)^(2*n+1)/factorial(2*n +1), n, 0, Infinity)
    
#by the same logic, we derive my_cosine as follows:
def my_cosine(x):
    n=var('n')
    return sum((-1)^n*(x)^(2*n)/factorial(2*n ), n, 0, Infinity)
    
show(cos(3.5).n())
show(my_cosine(3.5).n())
show(sin(1.56).n())
show(my_sine(1.56).n())
#plotting my_sine vs sin, my_cosine vs cos
phase = pi/2
#plot(cos(x), (x, 0, 2*pi), color='pink', aspect_ratio=1)+plot(my_cosine(x), (x, 0, 2*pi), color='blue', aspect_ratio=1)+plot(my_sine(x), (x, 0, 2*pi), color='red', aspect_ratio=1)+ plot(sin(x), (x, 0, 2*pi), color="yellow", aspect_ratio=1 )

#question: can we write cos(x-pi/2) into two parts, cos(x), and some constant?

show(cos(pi/2))

#plot(cos(x-phase), (x, 0, 2*pi), color='pink', aspect_ratio=1)+  plot(sin(x), (x, 0, 2*pi), color="yellow", aspect_ratio=1 )
plot(cos(x-phase), (x, 0, 2*pi), color='pink', aspect_ratio=1)+  plot(cos(x), (x, 0, 2*pi), color="yellow", aspect_ratio=1 )

#sin x= cos(x -pi/2)
```

```python
omega = pi/4
f(t) = sin(omega * t)
fplot = plot(f, (t, 0, 100))
f_prime(t) = f.diff(t)
f_double_prime(t) = f.diff(t, 2)
f_primeplot = plot(f_prime, (t, 0, 100), color="indigo")
f_doubleprimeplot = plot(f_double_prime, (t, 0, 100), color="red")
(fplot + f_primeplot + f_doubleprimeplot).show()
```
