This graph verifies that the so called double angle identify of sine function, then we can apply product rule to C2 to calculate the derivative as below:

```python
def Product_Rule(f, g, x):
  f_prime(x) = f.diff(x)
  g_prime(x) = g.diff(x)
  P_prime(x) = f(x)*g_prime(x) + g(x)*f_prime(x)
  return P_prime(x)
```

**Chain Rule:**
If $g$ is differentiable at $x$ and $f$ is differentiable at $g(x)$, then the composite function $C$ defined by $C(x) = f(g(x))$ is differentiable at $x$ and

$$C'(x) = f'(g(x))g'(x)$$



```python
def Chain_Rule(f, g, x):
  f_prime(x) = f.diff(x)
  g_prime(x) = g.diff(x)
  C_prime(x) = f_prime(g(x))*g_prime(x)
  return C_prime(x)
```

Verbal interpretation:

"If $C$ is a composite function defined by an outer function $f$ and an inner function $g$, then $C'$ is given by the derivative of the outer function evaluated at the inner function, then multiplied with the derivative of the inner function."

```python
f(x)=x^2
g(x)=tan(x)
r(x) = f(g(x))
show(r(x))
show(Chain_Rule(f,g,x))
```
