**Construction of wave model**

$\frac{\theta} {x} = \frac{2pi}{ \lambda} $

Solving it by:

```python
theta, x, llambda = var('theta', 'x', 'llambda')
show(solve([theta / x == 2 * pi / llambda], theta))
```

This matches the result in the text book mentioned above, hence we come to the following formulation of the wave function:

$y(x) = A \sin (\frac{2 \pi} {\lambda} x) $

and therefore with constant velocity v and time t:

$y(x, t) = A \sin(\frac{2 \pi} {\lambda} (x - vt))$

or factored out:

$y(x, t) = A \sin(\frac{2 \pi} {\lambda} x - \frac{2 pi} {\lambda}  vt))$

making $k = \frac{2 \pi} {\lambda}$ as the **wave number**, which has units of inverse meters, $m^{-1}$

and angular frequency $\omega = \frac{2 \pi} {T}$

[https://keylonticwiki.peertalk.net/index.php/Math_developing_angular_frequency_formular#now_we_are_ready_for_.5Bmath.5D.5Cdisplaystyle.7B_.5Comega_.7D.5B.2Fmath.5D https://keylonticwiki.peertalk.net/index.php/Math_developing_angular_frequency_formular#now_we_are_ready_for_.5Bmath.5D.5Cdisplaystyle.7B_.5Comega_.7D.5B.2Fmath.5D]

since the wave moves with a constanct wave speed of  $v = \frac {\lambda} {T}$, the second term becomes:

$\frac{2 \pi} {\lambda}  vt = \frac{2 \pi} {\lambda} \frac{\lambda} {T} t = \frac{2 \pi} {T} t = \omega t $

with $\omega = \frac{2 \pi} {T}$

$\omega = \frac{2 \pi} {T}$  now started to make sense, in giving the rate of change of angles in radians, within the periodic function cycle, measured in time units, so we see that angular frequency is measured in radian/unit time, in other words, it is the speed of rotation around a center, and measured in the angel in radians, which is related the arc along the circumference of the spin, if we need to calculate the distance or force, or any other physical properties resulted in the rotation, the angular frequency can be applied in those formulations.

Finally, we arrive to the following equation:

$y(x, t) = A \sin ( k x \mp \omega t)$

The value 

$(k x \mp \omega t  \varphi)$ is known as the phase of the wave, where $\varphi$ is the initial phase of the wave function.

Whether the temporal term $\omega t$ is negative or positive depends on the direction of the wave.
