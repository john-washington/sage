**Velocity and Acceleration of the Medium**

The wave speed is constant and represents the speed of the wave propagating through the mediaum, not the speed of the particles making up the medium. The particles of the medium oscilate around and equilibrium position as the wave propagates through the medium. In the case of the transeverse wave propagating in the x-direction, the particle oscillate up and down in the y-direction, perpendicular to the motion of the wave. The velocity of the particles of the mediaum is not constant, which means there is an acceleration. The velocity of the medium, which is perpendicular to the wave velocity in a transverse wave, can be found by taking the partial derivative of the position equation with respect to time. 

The partial derivative is found by taking the derivative of the function, treating all variables as constants, except for the variable in question. In the case of the partial derivative with respect to time t, the position x is treated as a constant. Although this may sound strange if you haven't seen it before, the object of this exercise is to find the transverse velocity at a point, so in this sense, the x-position is not chaning. We have

$$
\begin{align}
y(x, t) &= A \sin(kx - \omega t + \varphi ) \\
v_y(x, t) &= \frac{\partial y(x, t)} {\partial t} \\
          &= \frac{\partial} {\partial t} (A \sin (kx - \omega t + \varphi )) \\
          &= - A \omega \cos (kx - \omega t + \varphi) \\
          &= -v_{y \max} \cos (kx - \omega t + \varphi).
\end{align}
$$


The magnitude of the maximum velocity of the media is $ | v_{y_{\max}} | = A \omega$. This may look familliar from the Oscillations and a mass on a spring.

We can find the acceleration of the medium by taking the partial derivative of the velocity equation with respect to time,

$$
\begin{align}
a_y(x, t) &= \frac{\partial v_y}{\partial t} \\
         &= \frac{\partial} {\partial t}( -A \omega \cos (kx - \omega t + \varphi)) \\
         &= -A \omega^2 \sin(kx - \omega t + \varphi) \\
         &= -a_{y_max}\sin(kx -\omega t + \varphi).
\end{align}
$$

The magnitude of the maximum acceleration is $|a_{y_{\max}} | = A \omega^2$. The particle of the medium, or the mass elements, oscillate in simple harmonic motion for a mechanical wave.

**The Linear Wave Equation**

We have just determined the velocity of the medium at a position x by taking the partial derivative, with respect to time, of the position y. For a transverse wave, this velocity is perpendicular to the direction of propagation of the wave. we found the acceleration by taking the partial derivative, with respective to time, of the velocity, which is the second time derivative of the position:

$a_y(x, t) = \frac{\partial ^2 y(x, t)}{\partial t^2} = \frac{\partial^2}{\partial t^2} ( A \sin (kx - \omega t + \varphi)) = -A \omega^2 \sin(kx - \omega t + \varphi)$

Now consider the partial derivatives with respect to the other variable, the position x, holding the time constant. the first derivative is the slope of the wave at a point x at a time t,

$\text{slope} = \frac{\partial y(x, t)}{\partial x} = \frac{\partial}{\partial x} ( A \sin (kx - \omega t + \varphi)) = A k \cos(kx - \omega t + \varphi)$.

The second partial derivative expresses how the slope of the wave changes with respect to position-in other words, the curvature of the wave, where

$\text{curvature} = \frac{\partial^2y(x,t)}{\partial x^2} = \frac{\partial^2}{\partial^2x} (A \sin(kx - \omega t + \varphi)) = -A k^2 \sin(kx - \omega t + \varphi)$.
