**Introduction**

This chapter is to develop shortcut derivative rules that will help us bypass the limit defintion fo the derivative in order to quickly determine the formula for $f'(x)$ when we are given a formula for $f(x)$. In the preview Section, we learned the rule for power functions, that if $f(x) = x^n$, then $f'(x)= n x ^{n-1}$, and justify this in part due to results from differemt n-values when applying the limit definition of the derivative.

Also the rule for expoential functions, that if $a$ is a postive real number, and $f(x) = a^x$, then $f'(x) = a^x ln(a)$.

**Exponential Functions:** $a \in \mathbb{R}, f(x) = a^x, f'(x)=a^x ln(a)$

$$\frac{d}{dx}[e^x] = e^x ln(e) = e^x$$

```python
show(ln(4))
```

```python
show(ln(e))
```

```python
P(t) = 24 + 8* sin(t)
P_prime(t) = P.diff(t)
show(P_prime(2))
show(P_prime(2).n())
plot(P, (t, 0, 5))
```

On Jan 1, 2030, the instantaneous rate of change of population is -3.3 hundres per decade, which means the population is declining at a rate of 3.3 hundred per decade on Jan 1, 2030.

**Summary**

1. The derivative function of an exponential function $f(x)=a^x$ is itself another exponential function, possible scaled.

2. Special case $f(x)= 2^x$, the $\frac{d}{dx}{[2^x]} = 2^x ln(2)$

3. $\frac{d}{dx}sin(x) = cos(x), \frac{d}{dx}cos(x) = -sin(x)$
