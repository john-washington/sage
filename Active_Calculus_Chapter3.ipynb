**First Derivative Test:**

If p is a critical number of a continuous function f that is differentiable near p(except possibly at x = p), then f has a relative maximum at p if and only if f' changes sign from positive to negative at p, and f has a relative minimum at p if and only if f' changes sign from negative to possitive at p.

**Example 3.1** Let f be a function whose derivative is given by the formula $f'(x) = e^(-2x) (3 -x) (x+1)^2$. Determine all critical numbers of f and decide whether a relative maixum, relative minimum, or neither occurs at each.

**Solution** 
Since we already have f'(x) written in factored form, it is straightforward to find the the critical numbers of f. Since f'(x) is defined for all values of x, we need only determine where $f'(x) = 0$. from the equation

$e^{-2x} (3 -x) (x+1)^2 =0$

**Note: let's use sagemath to solve this:**


```python
f_prime(x) = e^(-2*x)*(3-x)*(x+1)^2
show(f_prime(x))
show(solve([f_prime(x) == 0], x))
```

This result matches what the text book tell us!

It follows that 3 and -1 are critical numbers of f. 

Next, to apply the first derivative test, we'd like to know the sign of f'(x) at inputs near the critical numbers. Because the critical numbers are the only locations at which f' can change sign, it follows that the sign of the derivatives is the same on each of the intervals created by the critical numbers: for instance, the sign of f' must be the same for every $x < -1$. We create a first derivative sign chart to summarize the sign of f' on the relevatnt intervals along with the corresponding behavior of f.

**Second Derivative Test:** 
If p is a critical number of a continuous function f such that $f'(p) =0$ and $f''(p) \neq 0$, then f has a relative maximum at p if and only if $f''(p) < 0$, and f has a relative minimum at p if and only if $f''(p) > 0$
