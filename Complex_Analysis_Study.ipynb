Polar complex plane 
Modulus and argument

An alternative option for coordinates in the complex plane is the polar coordinate system that uses the distance of the point z from the origin (O), and the angle subtended between the positive real axis and the line segment Oz in a counterclockwise sense. This leads to the polar form

${\displaystyle z=re^{i\varphi }=r(\cos \varphi +i\sin \varphi )}$

of a complex number, where r is the absolute value of z, and $\varphi is$ the argument of z.

The absolute value (or modulus or magnitude) of a complex number 
$z = x + yi$ is

${\displaystyle r=|z|={\sqrt {x^{2}+y^{2}}}.}$

Together, r and Ï† give another way of representing complex numbers, the polar form, as the combination of modulus and argument fully specify the position of a point on the plane. Recovering the original rectangular co-ordinates from the polar form is done by the formula called trigonometric form

${\displaystyle z=r(\cos \varphi +i\sin \varphi ).}$

Using Euler's formula this can be written as

$z = r e^{i \varphi}$ or $z = r \exp i \varphi$
