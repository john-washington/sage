In general, we make the following definition: for an object moving in a straight line whose position at time t is give
by the function s(t), the average velocity of the object on the interval from to =a to t= b, denoted $AV_[a, b]}$ is given by the formula

$$AV_{[a,b]} =\frac{s(b) - s(a)}{b-a}$$

Note: the units on $AV_{[a,b]}$ are "units of s per unit of t", e.g. miles per hour or feet per second

Informally, we define the instantaneous velocity of a moving object at time t =a to be the value that the average velocity appproaches as we take smaller and smaller intervals of time containing t =a to compute the average velocity. 

**Developing the idea of limit**

we observe that the average velocity of the object on the interval $[a, a+h]$ is

$$ AV_{[a, a+h]} = \frac{s(a +h) - s(a)}{h}$$

with the denominator being simply $h$ because $(a+h) -a =h$

It is important to note that we allow $h$ to be a small negative number, too, enabling us to investigate the average velocity of the moving object on intervals prior to $t =a$, as well as folloiwng $t =a$.

When $h < 0$, $AV_{[a, a+h]}$ measures the average velocity on the interval $[a+h, a]$

To attempt to find theÂ´instantaneous velocity at t=a, we investigate what happens as the value of h approaches zero.

**The Derivative of a Function at a Point**

**Definition 1.1** Givena function $f$, a fixed input $x=a$, and a real number $L$, we say that $f$ has limit $L$ as $x$ approaches a, and write

$$\lim_{x \rightarrow a} f(x) = L$$

**Definition 1.3** Let $f$ be a function and $x = a$ a value in the function's domain. 
We define the derivative of $f$ with respect to $x$ evaluated at $x = a$, denoted $f'(a)$, by the formula
$$ f'(a) = \lim_{h \rightarrow 0} \frac{f(a+h) - f(a)}{h}$$
provided this limit exists.

The derivative of $f$ at the value $x =a$ is defined as the limit of the average rate of change of $f$ on the interval $[a, a+h]$ as $h \rightarrow 0$. It is possible for this limit not to exist, so not every function has a derivative at every point.

We say that a function that has a derivative at $x =a$ is differentiable at $x =a$

The derivative is a generalization of the instantaneous velocity of a position function: when $y = s(t)$ is a position
function of a moving body,  $s'(a)$ tells us the instantaneous velocity of the body at time $t =a$.

Because the quantity $\frac{f(a+h) -f(a)}{h}$ represents the slope of the line through $(a, f(a))$ and $(a +h, f(a +h))$, when we compute the derivative we are taking the limit of a collectin of slopes of lines, and thus the derivative itself represents the slope of a particularly importan line.

$f(x) = \frac{4-x^2}{x+2}$

The numberator and denominator are each polynomial functions, which are among the most well-behaved functions.
Formally, they are continuous, in that the limit of the function at any point is equal to its function value.

Situation is more complicated when $x \rightarrow -2$, dur in part to the fact that $f(-2)$ is not defined. We call 0/0 an **indeterminate form**

For now, we simply say we have more work to do. Observe that:

$$\lim_{x \rightarrow -2} f(x) = \lim_{x \rightarrow -2} \frac{4-x^2}{x+2} = \lim_{x \rightarrow -2} \frac{(2-x)(2+x)}{x+2}$$

We conclude that

$$
\lim_{x \rightarrow -2} f(x) = \lim_{x \rightarrow -2} 2 -x = 4$$
